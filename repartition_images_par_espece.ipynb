# Répartition des images par espèce

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Chemin du dossier contenant les sous-dossiers d'espèces
base_path = "C:/cours_b3/mspr-ia/WildLens_MSPR/Mammiferes"

# Comptage des images par espèce
species_counts = {}
for species in os.listdir(base_path):
    species_dir = os.path.join(base_path, species)
    if os.path.isdir(species_dir):
        count = len([f for f in os.listdir(species_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
        species_counts[species] = count

# Création du DataFrame
df = pd.DataFrame(list(species_counts.items()), columns=['Espèce', "Nombre d'images"])
df = df.sort_values(by="Nombre d'images", ascending=False)
df
```

```python
# Affichage du graphique
plt.figure(figsize=(10, 6))
plt.bar(df['Espèce'], df["Nombre d'images"])
plt.xticks(rotation=45, ha='right')
plt.xlabel("Espèce")
plt.ylabel("Nombre d'images")
plt.title("Répartition des images par espèce")
plt.tight_layout()
plt.show()
```
